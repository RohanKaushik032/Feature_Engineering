```python
##1. Scatter Plot (Numeric - Numeric)
```

```python
# Assuming you have a DataFrame called 'tips' with columns 'total_bill', 'tip', 'sex', 'smoker', and 'size'
sns.scatterplot(x=tips['total_bill'], y=tips['tip'], hue=tips['sex'], style=tips['smoker'], size=tips['size'])

# Customize the plot if needed
plt.title("Scatterplot of Total Bill vs. Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.legend(title="Sex", loc="upper right")
plt.show()
```

```python
## 2. Bar plot (Numerical-Categorical) 
```

```python
titanic.sample()
```

```python
sns.barplot(x='Pclass', y='Age', hue='Sex', data=titanic)
plt.title(" Age by Pclass and Sex")
plt.xlabel("Pclass")
plt.ylabel("Mean Age")
plt.show()
```

```python
##3. Box Plot(Numerical - Categorical)
```

```python
# Assuming you have a DataFrame called 'titanic' with columns 'Sex', 'Age', and 'Survived'
sns.boxplot(x='Sex', y='Age', hue='Survived', data=titanic)

# Customize the plot if needed
plt.title("Box Plot of Age by Sex and Survival")
plt.xlabel("Sex")
plt.ylabel("Age")
plt.legend(title="Survived", loc="upper right")
plt.show()
```

```python

# Assuming you have a DataFrame called 'titanic' with columns 'Pclass', 'Survived'
# and you want to create a heatmap of the cross-tabulation between 'Pclass' and 'Survived'
cross_tab = pd.crosstab(titanic['Pclass'], titanic['Survived'])

# Create a heatmap
sns.heatmap(cross_tab)

# Customize the plot if needed
plt.title("Survival Count by Passenger Class")
plt.xlabel("Survived")
plt.ylabel("Pclass")
plt.show()
```
